## 1 - Carregar os arquivos e cadastro, investimento1 e investimento2 e mostrar a quantidade de linhas de cada um, além das informações desses dataframes

## 2 - Juntar os dataframes de investimento1 e investimento2 em um só e ver a quantidade de linhas

## 3 - Crie um novo dataframe eliminando as linhas duplicadas

```python
inv = inv.drop_duplicates('id')
display(inv)
```

## 4 - Junte o dataframe de investimento com o de cadastro
Dica: use a chave id

```python
df = inv.merge(cad, on='id')

print(df.shape)
display(df.head(3))
```

## 5 - Ordene o dataframe pelo id, de forma crescente e redefina o índice

```python
df = df.sort_values('id', ignore_index=True)

print(df.shape)
display(df.head(3))
```

## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
def na_resumo(df):
  total_na = df.isna().agg(['sum', lambda x: (100*(x.sum()/len(df))).round(1)]).T
  total_na = total_na[total_na['sum'] > 0].rename(columns={"sum":"qtd_NaN", "<lambda>":"pct_NaN"})
  display(total_na.T)
  print('df shape: ', df.shape, '\n')
  pass

print('1. Resumo NAs antes do tratamento:')
na_resumo(df)

print('\n2.Preenchendo Ativo01 ~ Ativo07 == NaN com 0 ...')
df = df.fillna({f'Ativo0{e}':0 for e in range(1, 8)})
print('Resumo NAs após etapa 2:')
na_resumo(df)

print('\n3.Preenchendo Nota1 == NaN com Nota2, se Nota2 != NaN:')
#df = df.fillna(value={'Nota1':0, 'Nota2':0})
df['Nota1'] = df['Nota1'].fillna(df['Nota2'])
print('Resumo NAs após etapa 3:')
na_resumo(df)

print('\n4.Preenchendo Nota2 == NaN com Nota1, se Nota1 != NaN:')
df['Nota2'] = df['Nota2'].fillna(df['Nota1'])
print('Resumo NAs após etapa 4:')
na_resumo(df)

print('\n5.Eliminando linhas com Nota1 e Nota2 == NaN, pois representam menos de 1% do total:')
df = df.dropna(subset=['Nota1','Nota2'])
print('Resumo NAs após etapa 5:')
na_resumo(df)
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50


```python
colunas_ativos = df.columns[df.columns.str.contains('Ativ')]
df['Total'] = df[colunas_ativos].sum(axis=1)
df['Media'] = df[['Nota1', 'Nota2']].mean(axis=1).round().astype('Int64')
df['Dominio'] = df['Email'].apply(lambda x: x.split('@')[1])
df = df.rename(columns={"Salario":"Renda"})
df['Renda'] = df['Renda'].astype(str).str.replace('$','', regex=True)
df['Sexo'] = df['Sexo'].map({'Masc':'Masculino', 'Fem':'Feminino'}, na_action='ignore')
def f(x):
  if x > 80: return 'promotor'
  if x < 50: return 'detrator'
  return 'neutro'
df['NPS'] = df['Media'].apply(f)

display(df[['Email','Dominio','Renda','Sexo','Total','Nota1','Nota2','Media','NPS']])
```

## 8 - Calcule os seguintes dados
- Calcule o total investido em cada coluna Ativo (tente usar o apply)
- Qual o menor valor investido em cada ativo?
- Mostre a média e desvio padrão de cada ativo
- Qual a idade do investidor mais velho?
- Qual a mediana da coluna Média?

```python
ativos = ['Ativo01','Ativo02','Ativo03','Ativo04','Ativo05','Ativo06','Ativo07']

print('\nTotal investido em cada Ativo:')
r81 = df[ativos].apply([lambda x: x.sum()])
r81.index = ['total investido']
display( r81 )

print('\n\nMenor valor investido em cada Ativo:')
display( df[ativos].agg(['min']) )

print('\n\nMédia/Desvio Padrão de cada Ativo:')
display( df[ativos].agg(['mean','std']) )

print('\n\nIdade do investidor mais velho:')
display( df['Idade'].max() )

print('\nMediana da coluna Média (das Notas 1 e 2):')
display( df['Media'].median() )
```

## 9 - Responder as perguntas abaixo

- Quantos investidores são promotores?
- Quantos investidores são do sexo feminino?
- Quantos investidores investiram mais de 16000 no Ativo04?
- Quantos investidores investiram mais de 16000 no Ativo04 e menos de 5000 no Ativo05?
- Quantos investidores do sexo feminino tem mais de 25 anos e investiu menos de 10000? (bonus: por que esse dado seria importante para a estratégia da nossa empresa?

- Quantos investidores tem em cada estado (dica: use groupby)
- Qual é a soma do total investido por cada estado?
- Como os investidores se distribuem em promotores, detratores e neutros?
- Qual é a média do total investido por promotores, detratores e neutros?


```python

print('\nQuantos investidores são promotores?')
r91 = df['id'][df['NPS'] == 'promotor'].count()
print(f'RESPOSTA: {r91} investidores são promotores')

print('\nQuantos investidores são do sexo feminino?')
r92 = df['id'][df['Sexo'] == 'Feminino'].count()
print(f'RESPOSTA: {r92} investidores são do sexo feminino')

print('\nQuantos investidores investiram mais de 16000 no Ativo04?')
r93 = df['id'][df['Ativo04'] > 16000].count()
print(f'RESPOSTA: {r93} investidores investiram mais de 16000 no Ativo04')

print('\nQuantos investidores investiram mais de 16000 no Ativo04 e menos de \
5000 no Ativo05?')
m94 = (df['Ativo04'] > 16000) & (df['Ativo05'] < 5000)
r94 = df['id'][m94].count()
print(f'RESPOSTA: {r94} investidores')

print('\nQuantos investidores do sexo feminino tem mais de 25 anos e investiu \
menos de 10000? \nbonus: por que esse dado seria importante para a \
estratégia da nossa empresa?')
m95 = (df['Sexo'] == 'Feminino') & (df['Idade'] > 25) & (df['Total'] < 10000)
r95 = df['id'][m95].count()
print(f'RESPOSTA: {r95} investidores. Esse dado é importante pois esses são \
clientes em potencial para investirem mais!!!')

print('\n\nQuantos investidores tem em cada estado?')
r96 = df.groupby('Estado')[['Estado']].count()
display(r96.T.iloc[:,:14])
display(r96.T.iloc[:,14:])

print('\n\nQual é a soma do total investido por cada estado?')
r97 = df.groupby('Estado')[['Total']].apply(lambda x: x.sum().astype(int))
display(r97.T.iloc[:,:14])
display(r97.T.iloc[:,14:])

print('\n\nComo os investidores se distribuem em promotores, detratores e neutros?')
r98 = df.groupby('NPS')['NPS'].agg([('qtd','count')])
display(r98)

print('\n\nQual é a média do total investido por promotores, detratores e neutros?')
r99 = df.groupby('NPS')['Total'].agg([('investimento médio','mean')])
display(r99)
```
